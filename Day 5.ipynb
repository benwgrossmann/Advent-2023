```python
filename = "Day 5 Input.txt"

with open(filename,'r') as f:
    lines = f.read().splitlines()
```

```python
from bisect import bisect

class RangeMap:
    def __init__(self):
        self.ranges = []
        
    def add_range(self, dest_start, source_start, range_len):
        tup = (source_start, dest_start, range_len)
        idx = bisect(self.ranges, tup)
        self.ranges.insert(idx, tup)
    
    def apply_map(self, source):
        idx = bisect(self.ranges, (source,0,0)) - 1
        if idx == -1:
            return source
        source_start, dest_start, range_len = self.ranges[idx]
        diff = source - source_start
        if diff <= range_len:
            return dest_start + diff
        return source
```

```python
# Part 1 solution

def build_map(idx):
    new_map = RangeMap()
    for line in lines[idx+1:]:
        if not line:
            break
        new_map.add_range(*map(int, line.split()))
    return new_map

def chain_maps(rangemaps, x):
    for rmap in rangemaps:
        x = rmap.apply_map(x)
    return x

headers = [
 'seed-to-soil map:',
 'soil-to-fertilizer map:',
 'fertilizer-to-water map:',
 'water-to-light map:',
 'light-to-temperature map:',
 'temperature-to-humidity map:',
 'humidity-to-location map:']

seeds = list(map(int, lines[0].split()[1:]))
rangemaps = []

for head in headers:
    idx = lines.index(head)
    rangemaps.append(build_map(idx))

locations = [chain_maps(rangemaps,x) for x in seeds]
for seed, loc in zip(seeds, locations):
    print(f"seed {seed}: location {loc}")
print(f"min location: {min(locations)}")
```
