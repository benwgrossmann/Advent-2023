```python
filename = "Day 8 Input.txt"

with open(filename,'r') as f:
    lines = f.read().splitlines()
```

```python
instructions = lines[0]
lr_dict = {}

for line in lines[2:]:
    key, tail = line.split(" = ")
    lr_dict[key] = (tail[1:4], tail[6:9])
```

```python
# part 1 solution

lr_idx = {'L':0, 'R':1}

def do_path(state = 'AAA'):
    n_steps = 0
    while True:
        for c in instructions:
            state = lr_dict[state][lr_idx[c]]
            n_steps += 1
            if state == 'ZZZ':
                return n_steps

print(do_path())
```

```python
#part 2 solution

def get_first(initial):
    state = initial
    num_steps = 0
    while True:
        for c in instructions:
            state = lr_dict[state][lr_idx[c]]
            num_steps += 1
            if state[-1] == 'Z':
                return num_steps
```

```python
from math import gcd
from functools import reduce

def lcm(a,b):
    return a*b//gcd(a,b)

Q = [state for state in lr_dict if state[-1] == 'A']
firsts = [get_first(state) for state in Q]
print(firsts)

print(reduce(lcm,firsts))
```
